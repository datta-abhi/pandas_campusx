````python
ipl_nonull = ipl[~ipl['WinningTeam'].isnull()]
ipl_nonull.shape
````

````python
# Make a dataframe of each team in IPL with details like -
# Team Name, Matches Played, Win%, Home Win%, Away Win%. Show sorted dataframe on Win%

teams = ipl_nonull['Team1'].unique()

d = {}

for t in teams:
    d[t] = {}
    
#     total matches
    nos_matches = ipl_nonull[(ipl_nonull['Team1']==t) | (ipl_nonull['Team2']==t)].shape[0]
    
    # winning matches
    wins = ipl_nonull[ipl_nonull['WinningTeam']==t].shape[0]
    win_percent = np.round(100*wins/nos_matches,1)
    
#     home and away matches
    nos_home = ipl_nonull[ipl_nonull['Team1']==t].shape[0]
    nos_away = ipl_nonull[ipl_nonull['Team2']==t].shape[0]
    
#     home and away wins
    
    home_wins = ipl_nonull[(ipl_nonull['WinningTeam']==t) & (ipl_nonull['Team1']==t)].shape[0]
    away_wins = ipl_nonull[(ipl_nonull['WinningTeam']==t) & (ipl_nonull['Team2']==t)].shape[0]
    
#   home and away win percent
    home_win_percent = np.round(100*home_wins/ nos_home,2)
    away_win_percent = np.round(100*away_wins/ nos_away,2)
    
    
    d[t]['total_matches'] = nos_matches
    d[t]['win_percent'] = win_percent
    d[t]['homewin_percent'] = home_win_percent
    d[t]['awaywin_percent'] = away_win_percent
    
d

````

````python
pd.DataFrame(d).T.reset_index().sort_values('win_percent',ascending = False)
````

###`Q-9:` IPL Point Table

Make a function `point_table` which take `season` as parameter and show points table in non-ascendng order of points and in ascending order of team name.

For winning - 2 Ponits;
For loosing - 0 Point
For not result both team gets 1 points.

Make dataframe which will have
`TeamName` `MatchesPlayed` `MatchesWon` `NoResult` `Points`
make `TeamName` as index.

```
season parametr should be one of these->
['2022', '2021', '2020/21', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2009/10', '2009', '2007/08']
```


Output of two Top 2 in season 2022
```
TeamName    MatchesPlayed	MatchesWon	NoResult	Points

Gujarat Titans	    16	   12	       0	     24
Rajasthan Royals	  17	   10	       0	     20

```

````python
def team_stats(df,t):
    
    d = {}
    
    
    # total nos of matches 
    nos_matches = df[(df['Team1']==t) | (df['Team2']==t)].shape[0]
    
    # winning matches
    wins = df[df['WinningTeam']==t].shape[0]
    
    # no result matches
    no_result = df[((df['Team1']==t) | (df['Team2']==t)) & (df['WinningTeam'].isnull())].shape[0]
    
#     loses
    losses = (nos_matches - wins - no_result)
    
#     points
    points = 2*wins + no_result
    
    d['MatchesPlayed'] = nos_matches
    d['MatchesWon'] = wins
    d['NoResult'] = no_result
    d['Points'] = points
    
    return d
````

````python
temp1 = ipl[ipl['Season'] == '2007/08']

team_stats(temp1,'Rajasthan Royals')
````

````python
def point_table(season):
    
    season_df = ipl[ipl['Season'] == season]
    
    teams = set(list(season_df['Team1'].unique()) + list(season_df['Team2'].unique()))
    
    season_dict = {}
    
    for team in teams:
        season_dict[team] = team_stats(season_df,team)
    
    point_df = pd.DataFrame(season_dict).T.sort_values('Points',ascending = False)
    
    return teams,point_df
````

###`Q-10:` IPL Point Table cont.
Extend the above IPL Point Table with an extra column as `SeasonPosition`

Team below top 4 after sorting on `Points` and then on `TeamName` Will have same `SeasonPosition` as there rank. use rank function.

Teams in Top four will have `SeasonPosition` as:
```
    'Winner' - Team won final
    'Runner' - Team lost Final
    3 - Losing Team in Qualifier2
    4 - Losing Team in Eliminator
```

For changing value of pariticular cell use `df.at[row_index, col_label] = value`

Output of two Top 2 in season 2022. Your result should have all teams.
```
TeamName    MatchesPlayed	MatchesWon	NoResult	Points   SeasonPosition

Gujarat Titans	    16	   12	       0	     24         Winner
Rajasthan Royals	  17	   10	       0	     20         Runner

```

Note: If you try to chnage value of view of any dataframe a warnig will be shown. To avoid it, make a copy of the dataframe you want to change in by `df.copy()`

````python
def season_position(season):
    _,point_df = point_table(season)
    point_df = point_df.reset_index().rename(columns = {'index':'TeamName'})
    point_df = point_df.sort_values(by = ['Points','TeamName'],ascending = [False,True])
    point_df['rank'] = point_df['Points'].rank(ascending = False,method = 'first').astype('object')
    
    
#     final winner and runner up
    season_final = ipl[(ipl['Season']==season) & (ipl['MatchNumber']=='Final')]
    winner = season_final['WinningTeam'][0]
    runner_up = season_final.apply(lambda x: x['Team2'] if x['Team1']== x['WinningTeam'] else x['Team1'],axis = 1)[0]
    
    point_df['rank'] = point_df.apply(lambda x: 'Winner' if x['TeamName']==winner else x['rank'],axis = 1)
    point_df['rank'] = point_df.apply(lambda x: 'Runner' if x['TeamName']==runner_up else x['rank'],axis = 1)
    
#     3rd --> Losing team in Qualifier 2
    qualifier2 = ipl[(ipl['Season']==season) & (ipl['MatchNumber']=='Qualifier 2')]
#     qualifier_winner = qualifier2['WinningTeam'][0]
    third = qualifier2.apply(lambda x: x['Team2'] if x['Team1']== x['WinningTeam'] else x['Team1'],axis = 1)[0]
    point_df['rank'] = point_df.apply(lambda x: 'Third' if x['TeamName']==third else x['rank'],axis = 1)


#     4th - Losing Team in Eliminator
    
    
    
    return winner,runner_up, point_df
````
